```python
import pandas as pd

def aggiorna_ospiti_con_dati_uniformati(file_principale, file_duplicati, file_output):
    df_principale = pd.read_csv(file_principale)
    df_duplicati = pd.read_csv(file_duplicati)

    dati_uniformati = {}
    for _, row in df_duplicati.iterrows():
        nome = row['Nome']
        if nome not in dati_uniformati:
            dati_uniformati[nome] = {
                'Nazionalità': row['Nazionalità'],
                'Sesso': row['Sesso'],
                'Età': row['Età'],
                'Professione': row['Professione'],
                'Notorietà': row['Notorietà'],
                'Numero di follower social': row['Numero di follower social']
            }
    
    
    df_aggiornato = df_principale.copy()
    ospiti_aggiornati = []
    
    for idx, row in df_aggiornato.iterrows():
        nome_ospite = row['Nome']
        
        if nome_ospite in dati_uniformati:
            dati = dati_uniformati[nome_ospite]
            
            df_aggiornato.at[idx, 'Nazionalità'] = dati['Nazionalità']
            df_aggiornato.at[idx, 'Sesso'] = dati['Sesso']
            df_aggiornato.at[idx, 'Età'] = dati['Età']
            df_aggiornato.at[idx, 'Professione'] = dati['Professione']
            df_aggiornato.at[idx, 'Notorietà'] = dati['Notorietà']
            df_aggiornato.at[idx, 'Numero di follower social'] = dati['Numero di follower social']
            
            if nome_ospite not in ospiti_aggiornati:
                ospiti_aggiornati.append(nome_ospite)
    
    df_aggiornato.to_csv(file_output, index=False)
    

    
   
    return df_aggiornato




file_principale = "../gem_data/ospiti_info_with_descriptions.csv"
file_duplicati = "../gem_data/ospiti_duplicati.csv" 
file_output = "../gem_data/ospiti_info_aggiornato.csv"

df_risultato = aggiorna_ospiti_con_dati_uniformati(
    file_principale, 
    file_duplicati, 
    file_output
)
    
```
