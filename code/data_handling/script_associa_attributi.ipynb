```python
import pandas as pd

ospiti_df = pd.read_csv('/Users/cristianrossato/Desktop/social/code_3_11/progetto/data/gemini_data/ospiti_cleared(6).csv')
associazione_df = pd.read_csv('/Users/cristianrossato/Desktop/social/code_3_11/progetto/data/guest_video_info_visual.csv')

nome_to_attributes = {}
for _, row in ospiti_df.iterrows():
    nome_to_attributes[row['nome']] = {
        'nome':row['nome'],
        'nazionalità': row['nazionalità'],
        'sesso': row['sesso'],
        'età': row['età'],
        'professione principale': row['professione principale'],
        'numero di follower social': row['numero di follower social'],
        'target audience': row['target audience']
    }

result_rows = []
for _, row in associazione_df.iterrows():
    nome = row['nome']
    if nome in nome_to_attributes:
        attrs = nome_to_attributes[nome]
        new_row = {
            'nome': attrs['nome'],
            'nazionalità': attrs['nazionalità'],
            'sesso': attrs['sesso'],
            'età': attrs['età'],
            'professione principale': attrs['professione principale'],
            'numero di follower social': attrs['numero di follower social'],
            'target audience': attrs['target audience'],
            'video_id': row['video_id'],
            'published_at': row['published_at'],
            'views': row['views'],
            'likes': row['likes'],
            'comments': row['comments']
        }
        result_rows.append(new_row)

result_df = pd.DataFrame(result_rows)

result_df = result_df.sort_values('published_at')
result_df.to_csv('/Users/cristianrossato/Desktop/social/code_3_11/progetto/data/associazione_guest_visual_with_attribute.csv', index=False)
```
